```python
import networkx as nx
from networkx.algorithms import tree
import time
import random
import matplotlib as plt
from collections import defaultdict
from heapq import heappush, heappop
import heapdict as hp
from itertools import count

def make_grid(size, g_type=nx.grid_2d_graph):
    """
    Makes a grid graph of size*size number of vertices.
    The graph has a uniform distribution of weights
    between 0 and 10.
    """
    G = g_type(size,size)
    G = nx.convert_node_labels_to_integers(G)
    for (u, v) in G.edges():
        G.edges[u,v]['weight'] = random.uniform(0,10)
    return G

def make_complete(size, g_type=nx.complete_graph):
    """
    Makes a complete graph with size number of vertices
    """
    G = g_type(size)
    G = nx.convert_node_labels_to_integers(G)
    for (u, v) in G.edges():
        G.edges[u,v]['weight'] = random.uniform(0,10)
    return G

def make_rand(size, prob, g_type=nx.gnp_random_graph):
    """
    Makes a random graph with a probabilty of edges
    equal to prob. """
    G = g_type(size, prob)
    G = nx.convert_node_labels_to_integers(G)
    for (u, v) in G.edges():
        G.edges[u,v]['weight'] = random.uniform(0,10)
    return G


G = make_grid(3)
```

```python
def dijkstra(graph, start):
    """
    Simple implementation of Dijkstra's algorithm.
    Uses an array to store the vertices and extracts
    them with a linear search
    """
    Q = set()
    dist = {}
    prev = {}
    
    for v in graph.nodes():
        dist[v] = float('inf')
        prev[v] = None
        Q.add(v)
    dist[start] = 0

    while Q:
        least = float('inf')
        for i in Q:
            pos = dist[i]
            if pos < least:
                least = pos
                u = i
        Q.remove(u)
        for v in graph[u]:
            alt = dist[u] + graph[u][v]['weight']
            if alt < dist[v]:
                dist[v] = alt
                prev[v] = u
            
    S = {}
    for i in prev:
        u = prev[i]
        S[i] = [i]
        while u != None:
            S[i].insert(0,u)
            u = prev[u]

    return S
```

```python
def dijkstra_heap(graph, start):
    """
    A dictionary heap implementation of Dijkstra's algorithm
    """
    Q = hp.heapdict()
    dist = {}
    prev = {}
    dist[start] = 0
    
    for v in graph.nodes():
        if v != start:
            dist[v] = float('inf')
        prev[v] = None
    Q[start] = 0

    while Q:
        u, cost = Q.popitem()
        for v in graph[u]:
            alt = dist[u] + graph[u][v]['weight']
            if alt < dist[v]:
                dist[v] = alt
                prev[v] = u
                Q[v] = alt
                
    S = {}
    for i in prev:
        u = prev[i]
        S[i] = [i]
        while u != None:
            S[i].insert(0,u)
            u = prev[u]
    return S
```

```python
def dijkstra_heap2(graph, start):
    """
    A binary heap implementation of Dijkstra's algorithm
    """
    push = heappush
    pop = heappop
    dist = {}
    seen = {}
    paths = defaultdict(list)
    c = count()
    # Q is a heap dictionary with node as key and (distance, c) values
    Q = []
    
    seen[start] = 0
    push(Q, (0, next(c), start))
    
    while Q:
        (d, _, v) = pop(Q)
        if v in dist:
            continue
        dist[v] = d
        for u, w in graph[v].items():
            potential = w['weight']
            vu_dist = dist[v] + potential
            if u not in seen or vu_dist < seen[u]:
                seen[u] = vu_dist
                push(Q,(vu_dist, next(c), u))
                paths[u] = paths[v] + [u]
                
    for i in paths:
        paths[i] = [start] + paths[i]
    
    if paths:
        return paths
    else:
        paths[start] = [start]
        return paths
```

```python
def dijkstra_lib(G,source):
    """
    Handle to the Networkx Dijkstra's algorithm implementation
    """
    return nx.single_source_dijkstra_path(G, source)

def prims_lib(G, source=0):
    """
    Handle to the networkx Prim's algorithm implementation
    """
    return list(tree.minimum_spanning_edges(G, algorithm='prim', data=False))  
```

```python
def prims(graph, source=0):
    """
    Simple implementatoin of Prim's algorithm that uses
    an array to store the vertices in and performs a 
    linear search to find the mimimum
    """
    ReachSet = []
    UnReachSet = {}
    for v in graph:
        UnReachSet[v] = (float("inf"), None)
    UnReachSet[source] = (0, None)
    SpanningTree = []
    
    while UnReachSet:
        smallest = float("inf")
        for i in UnReachSet:
            if UnReachSet[i][0] < smallest:
                smallest = UnReachSet[i][0]
                v = i
                last = UnReachSet[i][1]
        UnReachSet.pop(v)
                
        if v not in ReachSet:
            ReachSet.append(v)
            for x in graph[v]:
                if x not in ReachSet and UnReachSet[x][0] > graph[v][x]["weight"]:
                    UnReachSet[x] = (graph[v][x]["weight"], v)
            SpanningTree.append((last, v))

    return SpanningTree[1:]
```

```python
def prims(graph,source=0):
    ReachSet = [source]
    UnReachSet = list(range(1,len(graph)))
    SpanningTree = [source]
    
    while len(UnReachSet) > 0:
        smallest = float("inf")
        n = 0
        for i in ReachSet:
            for j in graph[i]:
                if j in ReachSet:
                    continue
                if graph[i][j]['weight'] < smallest:
                    smallest = graph[i][j]['weight']
                    n = j
        SpanningTree.append(n)
        ReachSet.append(n)
        UnReachSet.remove(n)
    return SpanningTree
```

```python
def prims_heap(graph, source=0):
    """
    An implementation of Prim's algorithm that uses
    a dictionary heap to store the vertices.
    """
    ReachSet = []
    UnReachSet = hp.heapdict()
    for v in graph:
        UnReachSet[v] = (float("inf"), None)
    UnReachSet[source] = (0, None)
    SpanningTree = []
    
    while UnReachSet:
        v, (cost, last) = UnReachSet.popitem()
        if v not in ReachSet:
            ReachSet.append(v)
            for x in graph[v]:
                if x not in ReachSet and UnReachSet[x][0] > graph[v][x]["weight"]:
                    UnReachSet[x] = (graph[v][x]["weight"], v)
            SpanningTree.append((last, v))

    return SpanningTree[1:]
```

```python
def data(exp, algorithm, g_type=make_grid):
    """
    A function that makes the data to determine the time complexity
    of algorithms. It runs the algorithm given in a loop
    exp times. Each loop makes a new graph and times the time it
    takes the algorithm to solve the problem. It then returns
    a list of times that can be ploted.
    """
    times = [0] * (exp + 1)
    for j in range(1):
        for i in range(1,exp+1):
            G = g_type(i)
            start = time.time()
            algorithm(G, 0)
            end = time.time()
            times[i] +=(end-start)
    times = [x / 10 for x in times]
    return times

def data_edge(exp, algorithm, prob=.01):
    """
    A function that makes the data to determine the time complexity
    of the algorithm when the number of edges increase. There are two
    loops one runs the experiment 10 times, the other increases the
    probability of edges by .01 * i where i is the number of iterations.
    The probability of edges is from .01 to 1, sparse to fully connected.
    The graph is an Erdős-Rényi graph.
    """
    times = [0]*50
    for j in range(10):
        for i in range(1,50):
            G = make_rand(exp, prob * i)
            start = time.time()
            algorithm(G, 1)
            end = time.time()
            times[i] += (end-start)
    times = [x / 10 for x in times]
    return times

def test_algorithms(exp, algorithm, algorithm2):
    for i in range(1,exp+1):
        G = make_grid(i)
        result1 = algorithm(G, 0)
        result2 = algorithm2(G, 0)
        if result1 != result2:
            print(result1)
            print(result2)
        assert(result1 == result2)
        
    print("DONE")
```

```python
# Tests to check that the algorithms are correct

test_algorithms(30,prims_lib, prims)
test_algorithms(30,prims_lib, prims_heap)
test_algorithms(30,dijkstra_lib, dijkstra)
test_algorithms(30,dijkstra_lib, dijkstra_heap)
test_algorithms(30,dijkstra_lib, dijkstra_heap2)
```
